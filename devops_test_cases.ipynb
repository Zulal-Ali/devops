```python
#TEST CASE 1: Customer Registration
from selenium import webdriver
from selenium.webdriver.common.by import By
import time

# Set up the WebDriver
driver = webdriver.Chrome()

try:
    # Open the website
    driver.get("http://localhost/ecommerce-app-h/customer_registration.php?register=1")

    # Locate the sign-up form elements
    
    f_name_input = driver.find_element(By.ID,"f_name")
    l_name_input = driver.find_element(By.ID,"l_name")
    email_input = driver.find_element(By.ID,"email")
    password_input = driver.find_element(By.ID,"password")
    repassword_input = driver.find_element(By.ID,"repassword")
    mobile_input = driver.find_element(By.ID,"mobile")
    address1_input = driver.find_element(By.ID,"address1")
    address2_input = driver.find_element(By.ID,"address2")
    signup_button = driver.find_element(By.NAME,"signup_button")


    # Enter valid sign-up credentials
    f_name_input.send_keys("CustomerFirstName")
    l_name_input.send_keys("CustomerLastName")
    email_input.send_keys("customer_email@example.com")
    password_input.send_keys("customer_secure_password")
    repassword_input.send_keys("customer_secure_password")
    mobile_input.send_keys("1234567890")
    address1_input.send_keys("123 Main Street")
    address2_input.send_keys("Apt 45")

    # Click on the sign-up button
    signup_button.click()

    # Wait for a moment to let the registration process complete
    time.sleep(2)

    error_message_div = driver.find_element(By.CLASS_NAME,"panel-footer")
    assert not error_message_div.text.strip(), "Signup failed."
    
    print("Successful")

except Exception as e:
    print(f"Test failed: {e}")

finally:
    # Close the browser window
    driver.quit()
```

```python
#TEST CASE 3: Add Product to Cart
from selenium import webdriver
from selenium.webdriver.common.by import By
import time

# Set up the WebDriver
driver = webdriver.Chrome()

try:
    # Open the website
    driver.get("http://localhost/ecommerce-app-h/profile.php")

    # Navigate to the product listing page
    driver.find_element(By.LINK_TEXT,"Products").click()

    # Select a product and click on "Add to Cart" button
    product_link = driver.find_element(By.XPATH,"//div[@class='product']//a")
    product_link.click()

    
    add_to_cart_button = driver.find_element(By.ID,"add_to_cart_button")
    add_to_cart_button.click()

    time.sleep(2)

    cart_icon = driver.find_element(By.ID,"cart_icon")
    assert cart_icon.text == "1", "Adding product to cart failed."

    driver.find_element(By.LINK_TEXT,"Cart").click()

    # Verify that the selected product is listed in the cart
    cart_product = driver.find_element(By.XPATH,"//div[@class='cart-product']")
    assert cart_product.is_displayed(), "Product not found in the cart."

    print("Test Case 1: Add Product to Cart - Passed")

except Exception as e:
    print(f"Test Case 1: Add Product to Cart - Failed: {e}")

finally:
    # Close the browser window
    driver.quit()
```

```python
#TEST CASE 4: Search for a Product
from selenium import webdriver
from selenium.webdriver.common.by import By
# Set up the WebDriver
driver = webdriver.Chrome()

try:
    # Open the website
    driver.get("http://localhost/ecommerce-app-h/index.php")

    # Enter a search query and click on the search button
    search_input = driver.find_element(By.ID,"search")
    search_input.send_keys("your_search_query")
    search_button = driver.find_element(By.ID,"search_btn")
    search_button.click()

    assert "Search Results" in driver.title, "Search results page not displayed."

    search_result = driver.find_element(By.XPATH,"//div[@class='search-result']//a")
    search_result.click()

    assert "Product Details" in driver.title, "Product details page not displayed."

    print("Test Case 2: Search for a Product - Passed")

except Exception as e:
    print(f"Test Case 2: Search for a Product - Failed: {e}")

finally:
    driver.quit()
```

```python
#TEST CASE 5: Checkout Process
from selenium import webdriver
from selenium.webdriver.common.by import By
import time

# Set up the WebDriver
driver = webdriver.Chrome()

try:
    # Open the website
    driver.get("http://localhost/ecommerce-app-h/index.php")

    driver.find_element_by_link_text("Cart").click()
    checkout_button = driver.find_element(By.ID,"proceed_to_checkout_button")
    checkout_button.click()

    # Fill in shipping and payment information
    shipping_input = driver.find_element(By.ID,"shipping_address")
    shipping_input.send_keys("123 Main St")

    payment_input = driver.find_element(By.ID,"credit_card")
    payment_input.send_keys("1234 5678 9012 3456")

    # Click on the "Place Order" button
    place_order_button = driver.find_element(By.ID,"place_order_button")
    place_order_button.click()

    time.sleep(2)

    assert "Order Confirmation" in driver.title, "Order confirmation page not displayed."

    print("Test Case 3: Checkout Process - Passed")

except Exception as e:
    print(f"Test Case 3: Checkout Process - Failed: {e}")

finally:
    driver.quit()
```
